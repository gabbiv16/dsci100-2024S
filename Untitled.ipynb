```r
# read in the libraries

library(tidyverse)
library(tidymodels)
library(repr)
library(RColorBrewer)
options(repr.matrix.max.rows = 10)
```

```r
# read the dataset and write a short description.
cleveland_data <- read_csv("data/heart_disease/processed.cleveland.data", col_names = FALSE)

#rename the columns
cleveland_data <- rename(cleveland_data, age = X1, sex = X2, cp = X3, trestbps = X4, chol = X5, fbs = X6, restecg = X7, thalach = X8, exang = X9, oldpeak = X10, slope = X11, ca = X12, thal = X13, num = X14)

#select the relevant columns (age, chol, trestbps, num)
cleveland_selected <- select(cleveland_data, age, chol, trestbps, num) |> 
    arrange(age)

#wrangle the data, set num to either True/False
cleveland_wrangled <- cleveland_selected |>
    mutate(num = as_factor(num)) |>
    mutate(num = fct_recode(num, "False" = "0", "True" = "1", "True" = "2", "True" = "3", "True" = "4"))
cleveland_wrangled

#split dataset into training + testing
cleveland_split <- initial_split(cleveland_wrangled, prop = 0.75, strata = num)
cleveland_train <- training(cleveland_split)
cleveland_test <- testing(cleveland_split)

head(cleveland_train)
```

```r
#predictive question: how do age, chol(cholesterol) and trestbps (resting blood pressure) predictors predict the num (or diagnosis of heart disease)?

#use group_by & summarize to calculate the min, max + mean for age, chol and trestbps predictors.
cleveland_summarize <- cleveland_train |>
    group_by(num) |>
    summarize(min_age = min(age, na.rm = TRUE), max_age = max(age, na.rm = TRUE), mean_age = mean(age, na.rm = TRUE), min_chol = min(chol, na.rm = TRUE), max_chol = max(chol, na.rm = TRUE), 
              mean_chol = mean(chol, na.rm = TRUE), min_trestbps = min(trestbps, na.rm = TRUE), max_trestbps = max(trestbps, na.rm = TRUE), mean_trestbps = mean(trestbps, na.rm = TRUE)
              )
cleveland_summarize
              
```

```r
#plotting.
# plot 1: age + chol, plot 2: age + bp, plot 3: chol + bp
set.seed(1234)

cleveland_plot_1 <- ggplot(cleveland_train, aes(x = age, y = chol, color = num)) + 
    geom_point() +
    labs(x = "Age", y = "Cholesterol (mg/dl)", color = "Diagnosis")
cleveland_plot_1

cleveland_plot_2 <- ggplot(cleveland_train, aes(x = age, y = trestbps, color = num)) +
    geom_point() +
    labs(x = "Age", y = "Resting Blood Pressure (in mm Hg)", color = "Diagnosis")
cleveland_plot_2

cleveland_plot_3 <- ggplot(cleveland_train, aes(x = chol, y = trestbps, color = num)) +
    geom_point() +
    labs(x = "Cholesterol (mg/dl)", y = "Resting Blood Pressure (in mm Hg)", color = "Diagnosis")
cleveland_plot_3

 
```
